```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("train.csv")
data.head()
```

```python
print(data['Cover_Type'].value_counts())
sns.countplot(x="Cover_Type", data=data)
```

```python
soils = data.iloc[:,15:55]
values = []
for col in soils.columns:
    if(soils[col].value_counts()[0] == len(soils)):
        values.append(0)
        #print(col, 0)
    else:
        values.append(soils[col].value_counts()[1])
        #print(col, soils[col].value_counts()[1])
print(sum(values))     
        

def plot_bar_x(label, values):
    # this is for plotting purpose
    
    index = np.arange(len(label))
    plt.figure(figsize = (10, 5))
    plt.bar(index, values)
    plt.xlabel('Soil Type', fontsize=10)
    plt.ylabel('Occurrences', fontsize=10)
    plt.xticks(index, label, fontsize=6, rotation=30)
    #plt.title('Número de objetos para cada tipo de solo')
    
    plt.show()
    
plot_bar_x(soils.columns, values)
```
