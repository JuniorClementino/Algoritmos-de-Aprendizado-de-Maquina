```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  
%matplotlib inline

import scipy.cluster.hierarchy as shc
from sklearn.cluster import KMeans

import time
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler, StandardScaler

pd.options.display.max_colwidth = 150

data_iris = pd.read_csv("iris.csv")
data_glass = pd.read_csv("glass.data")

print('Iris:', np.shape(data_iris))
print('Glass:', np.shape(data_glass))
```

```python
data_glass = shuffle(data_glass)
data_glass.head()
```

```python
data_iris = shuffle(data_iris)
data_iris.head()
```

```python
plt.figure(figsize=(18, 7))  
plt.title("Dendograms_Iris")  
dend = shc.dendrogram(shc.linkage(data_iris.iloc[:,0:4], method='ward'))  
```

```python
plt.figure(figsize=(18, 7))  
plt.title("Dendograms_Glass")  
dend = shc.dendrogram(shc.linkage(data_glass.iloc[:,:], method='ward')) 
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(data_iris.iloc[:,0:4])
centers = kmeans.cluster_centers_
y_kmeans = kmeans.predict(data_iris.iloc[:,0:4])

plt.scatter(data_iris.iloc[:, 0], data_iris.iloc[:, 1], c=y_kmeans, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

```python
kmeans = KMeans(n_clusters=6)
kmeans.fit(data_glass.iloc[:,:])
centers = kmeans.cluster_centers_
y_kmeans = kmeans.predict(data_glass.iloc[:,:])

plt.scatter(data_glass.iloc[:, 0], data_glass.iloc[:, 1], c=y_kmeans, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```
